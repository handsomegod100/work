```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)
X , Y = np.meshgrid(x,y)
XX = np.c_[X.ravel(),Y.ravel()]
```

```python
def f(x): 
    c=[]
    for i in x:
        fxy = 1/(8*(2*sp.pi))*np.exp((-1/2)*((i[0]/4)**2+(i[1]/2)**2))
        c.append(fxy)
    c = np.array(c)
    return c
c = f(XX)
print(c)
```

```python
c = c.reshape(X.shape)
```

```python
plt.contourf(X,Y,c,cmap="RdYlBu",alpha=0.2)
```
