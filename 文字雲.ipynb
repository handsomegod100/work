```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba
from pyquery import PyQuery as pq
```

```python
html = pq(url="https://tw.news.yahoo.com/%E4%B8%80%E8%8A%B3%E6%B0%B4%E6%9E%9C%E8%8C%B6%E5%8F%88-1-%E5%AE%B6%E6%94%B6%E6%94%A4%E8%99%8E%E5%B0%BE%E4%B8%AD%E6%AD%A3%E5%BA%97%E4%B8%8B%E5%8F%B0%E4%B8%80%E9%9E%A0%E8%BA%AC-003638635.html"
          ,encoding='utf-8')
print(html)
```

```python
print(html('.caas-body').text())
word = html('.caas-body').text()
s = word.find("更多相關新聞")
print(word[:s])
newword = word[:s]
```

```python
jieba.set_dictionary('/Users/mingshengchen/anaconda3/lib/python3.7/site-packages/jieba/dict.txt.big.txt')
jieba.add_word('一芳')
wordlist = jieba.cut(newword , cut_all = False)
```

```python
wl_space_split = " ".join(wordlist)
wl_space_split
```

```python
wc = WordCloud(font_path="/System/Library/Fonts/STHeiti Light.ttc", #設置字體
               background_color="white", 
               width=1000,height=600,
               min_font_size=20,
               max_words=200,
               max_font_size=80 )    
wc.generate(wl_space_split)
```

```python
plt.axis("off")
plt.figure(figsize=(12,20),dpi=300)
plt.imshow(wc) 
plt.show()
```
