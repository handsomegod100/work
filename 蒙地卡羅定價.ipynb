```python
import pandas as pd
import numpy as np
import math
```

```python
x = np.random.normal(0,1,size =(1000000,1)) #隨機變數
s = float(input("請輸入股價："))
k = float(input("請輸入履約價："))
t = float(input("請輸入到期年數："))          
n = float(input("請輸入期數："))
r = float(input("請輸入無風險利率(小數)："))
vix = float(input("請輸入股價波動率(小數)："))
dt = t/n
```

```python
Frame = pd.DataFrame(x,columns = ["隨機變數"]) #生成表格
Frame
```

```python
for i in range(Frame.shape[0]): #計算到期股價
    Frame.loc[i,"到期股價"] = s * math.exp((r-(0.5*(vix**2)))*dt+vix*(-Frame.iloc[i,0])*(dt))
Frame.head()
```

```python
for i in range(Frame.shape[0]): #計算到期買權價值
    Frame.loc[i,"到期買權價值"] = max(0,Frame.iloc[i,1]-k)
Frame.head()
```

```python
Frame["到期買權價值"].describe()
```

```python
Frame["到期買權價值"].mean()/(1+r) #買權現值
```
