```python
from pyquery import PyQuery as pq
from ipywidgets import interact
```

```python
html = pq("http://rate.bot.com.tw/xrt?Lang=zh-TW")
print(html)
```

```python
currencies = html('td[data-table="本行現金賣出"].rate-content-cash').text()
print(currencies)
currencies = html('td[data-table="本行現金賣出"].rate-content-cash').text().split()
```

```python
currency_names = html('div.hidden-phone').text()
print(currency_names)
currency_names = html('div.hidden-phone').text().split()
```

```python
currency_dict = {}
currency_idx = 0
for idx , currency_name in enumerate(currency_names):
    if idx % 2 == 0:
        currency_dict[currency_name] = currencies [currency_idx]
        currency_idx +=1
print(currency_dict)
```

```python
def f(幣種):
    html = pq("http://rate.bot.com.tw/xrt?Lang=zh-TW")
    currencies = html('td[data-table="本行現金賣出"].rate-content-cash').text().split()
    currency_names = html('div.hidden-phone').text().split()
    currency_dict = {}
    currency_idx = 0
    for idx , currency_name in enumerate(currency_names):
        if idx % 2 == 0:
            currency_dict[currency_name] = currencies [currency_idx]
            currency_idx +=1
    print('現在台幣兌換{}的匯率是{}'.format(幣種, currency_dict[幣種]))
interact(f, 幣種=currency_dict.keys())
```
